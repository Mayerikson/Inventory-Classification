# 1 An√°lise da MYK Supply

## 1.1 Cen√°rio: Dominando a Intelig√™ncia de Estoque com Dados

Nos √∫ltimos trimestres, a MYK Supply, empresa refer√™ncia em distribui√ß√£o de produtos para o setor de tecnologia e consumo, vem enfrentando desafios operacionais cr√≠ticos relacionados √† gest√£o de estoque. Em um mesmo armaz√©m, itens essenciais sofrem com ruptura de estoque, enquanto outros, com baixa rotatividade, ocupam espa√ßo valioso e geram custos ocultos.

A √°rea financeira e o time de log√≠stica pressionam por mais previsibilidade e efici√™ncia, apontando para a necessidade de decis√µes estrat√©gicas baseadas em dados.

Diante desse cen√°rio, a lideran√ßa executiva foi direta:

*‚ÄúPrecisamos saber quais produtos merecem aten√ß√£o total e quais est√£o drenando recursos. A gest√£o de estoque precisa ser guiada por dados e prioriza√ß√£o inteligente.‚Äù*

A partir dessa demanda, o time de analytics iniciou um projeto de intelig√™ncia operacional, com um objetivo claro: **transformar dados hist√≥ricos de consumo em a√ß√µes que reduzam custos e aumentem o ROI log√≠stico.**

## 1.2 Objeto do Projeto

Otimizar a gest√£o de estoques da MYK Supply por meio da classifica√ß√£o ABC-XYZ, visando priorizar itens de maior impacto financeiro e operacional, reduzir capital imobilizado e ampliar a previsibilidade nas decis√µes de compra e reposi√ß√£o.

Para isso, o time de dados buscar√° responder as seguintes perguntas:

1)Quais itens representam maior valor financeiro e devem receber aten√ß√£o m√°xima (Classe A)?

2)Quais itens t√™m consumo mais imprevis√≠vel e exigem buffers ou estrat√©gias diferenciadas (Classe Z)?

3)Quais combina√ß√µes ABC-XYZ orientam pol√≠ticas espec√≠ficas de reabastecimento?

4)Qual o potencial de economia ao redirecionar recursos para os grupos mais estrat√©gicos (ex: AX)?

5)Quais recomenda√ß√µes pr√°ticas podem ser entregues para as √°reas de compras, log√≠stica e financeiro com base nessa prioriza√ß√£o?

## 1.3 Premissas do Projeto

Este estudo tem como foco principal a an√°lise da performance e previsibilidade dos itens em estoque, reconhecendo que a aloca√ß√£o inteligente de recursos log√≠sticos √© um vetor direto de ROI.

Embora o retorno financeiro ainda n√£o tenha sido mensurado de forma direta nesta fase, os padr√µes identificados aqui,como o foco em produtos AX e a poss√≠vel descontinua√ß√£o de itens CZ e servir√£o como base para simula√ß√µes financeiras futuras e acompanhamento de KPIs de redu√ß√£o de custo e giro de estoque.

A literatura de Supply Chain e experi√™ncias anteriores mostram que empresas que priorizam itens cr√≠ticos (alto valor e alta previsibilidade) conseguem:

Assim, este projeto prepara o terreno para uma segunda etapa mais voltada √† mensura√ß√£o do ROI e √† automa√ß√£o de pol√≠ticas de abastecimento.


**O que √© classifica√ß√£o ABC-XYZ?**

√â uma t√©cnica de gest√£o de estoques que combina dois crit√©rios: o valor financeiro dos itens (ABC) e a previsibilidade de consumo (XYZ). A an√°lise ABC identifica quais produtos s√£o mais valiosos para o neg√≥cio (Classe A = alto impacto financeiro; C = baixo), enquanto a XYZ avalia a regularidade da demanda (X = consumo est√°vel; Z = imprevis√≠vel). Ao cruzar essas duas dimens√µes, cria-se uma matriz com 9 categorias (como AX, BY, CZ), permitindo priorizar recursos, definir estrat√©gias de reabastecimento sob medida e melhorar o ROI log√≠stico.

```python
# Importa√ß√£o das bibliotecas necess√°rias

import pandas as pd
import warnings


arquivo = "abc_xyz_dataset.csv"

# Suprimir todos os avisos

warnings.filterwarnings ("ignore")
# : Importar o dataset 
df = pd.read_csv(arquivo)

# 1. Printar a contagem do n√∫mero de linhas e colunas
print("‚û°Ô∏è N√∫mero de linhas e colunas:", df.shape)

# 2. Printar a contagem do n√∫mero de valores missing por vari√°vel
print("\n‚û°Ô∏è Valores faltantes por vari√°vel:")
print(df.isna().sum())

# 3. Printar a contagem de linhas duplicadas
duplicadas = df.duplicated().sum()
print(f"\n‚û°Ô∏è N√∫mero de linhas duplicadas: {duplicadas}")

# 4. Mostrar os tipos de dados usando o comando dtypes
print("\n‚û°Ô∏è Tipos de dados das vari√°veis:")
print(df.dtypes)

# 5. Printar uma amostra aleat√≥ria de 30 linhas com o comando .sample()
print("\n‚û°Ô∏è Amostra aleat√≥ria de 30 linhas:")
print(df.sample(30))

# 6. Mostrar as primeiras 5 linhas com .head()
print("\n‚û°Ô∏è Primeiras 5 linhas do dataset:")
print(df.head())
```

**Interpreta√ß√£o:**

Ao verificar os tipos de dados, percebemos que:

As vari√°veis "Price_Per_Unit" e "Total_Sales_Value" podem ser do tipo float, especialmente se houver valores decimais.

Para garantir que tudo corra bem, vamos adicionar um trecho de c√≥digo para converter essas colunas para float, se necess√°rio. Isso ajuda a evitar problemas futuros nos c√°lculos financeiros!




```python
# Converter tipos espec√≠ficos, se necess√°rio
df['Price_Per_Unit'] = df['Price_Per_Unit'].astype(float)
df['Total_Sales_Value'] = df['Total_Sales_Value'].astype(float)

# Mostrar novamente os tipos de dados atualizados
print("\n‚û°Ô∏è Tipos de dados atualizados:")
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from math import ceil
import warnings
warnings.filterwarnings('ignore')

# Configura√ß√µes de estilo
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi'] = 300

# Definindo vari√°veis quantitativas e qualitativas
var_quantitativas = ['Jan_Demand', 'Feb_Demand', 'Mar_Demand', 'Apr_Demand', 'May_Demand',
                     'Jun_Demand', 'Jul_Demand', 'Aug_Demand', 'Sep_Demand', 'Oct_Demand',
                     'Nov_Demand', 'Dec_Demand', 'Total_Annual_Units', 'Price_Per_Unit',
                     'Total_Sales_Value']

var_qualitativas = ['Item_ID', 'Item_Name', 'Category']

print("üìä AN√ÅLISE EXPLORAT√ìRIA DE DADOS")
print("=" * 60)
print(f"üî¢ Vari√°veis Quantitativas: {len(var_quantitativas)}")
print(f"üè∑Ô∏è Vari√°veis Qualitativas: {len(var_qualitativas)}")
print(f"üìã Shape do DataFrame: {df.shape}")
print("=" * 60)

# =============================================================================
# 1. AN√ÅLISE DE VARI√ÅVEIS QUALITATIVAS
# =============================================================================
print("\nüîç VALORES √öNICOS NAS VARI√ÅVEIS QUALITATIVAS:")
print("-" * 50)

for var in var_qualitativas:
    unique_vals = df[var].unique()
    print(f"\nüìã {var}:")
    print(f"   ‚Ä¢ Valores √∫nicos: {len(unique_vals)}")
    print(f"   ‚Ä¢ Valores nulos: {df[var].isnull().sum()}")
    
    # Mostrar apenas os primeiros 10 valores se houver muitos
    if len(unique_vals) <= 10:
        print(f"   ‚Ä¢ Valores: {list(unique_vals)}")
    else:
        print(f"   ‚Ä¢ Primeiros 10 valores: {list(unique_vals[:10])}")
        print(f"   ‚Ä¢ ... e mais {len(unique_vals) - 10} valores")

# =============================================================================
# 2. ESTAT√çSTICAS DESCRITIVAS QUANTITATIVAS
# =============================================================================
print("\n\nüìà ESTAT√çSTICAS DESCRITIVAS DAS VARI√ÅVEIS QUANTITATIVAS:")
print("-" * 50)

# Verifica se as colunas existem no DataFrame
var_quantitativas_existentes = [col for col in var_quantitativas if col in df.columns]

if var_quantitativas_existentes:
    stats_desc = df[var_quantitativas_existentes].describe().round(2)
    print(stats_desc)
    
    # Informa√ß√µes sobre valores nulos
    print("\nüîç VALORES NULOS:")
    null_info = df[var_quantitativas_existentes].isnull().sum()
    if null_info.sum() > 0:
        print(null_info[null_info > 0])
    else:
        print("‚úÖ Nenhum valor nulo encontrado!")
else:
    print("‚ùå Nenhuma vari√°vel quantitativa encontrada no DataFrame!")

# =============================================================================
# 3. VISUALIZA√á√ïES DAS VARI√ÅVEIS QUANTITATIVAS
# =============================================================================
print("\n\nüìä GERANDO VISUALIZA√á√ïES PARA VARI√ÅVEIS QUANTITATIVAS...")

if var_quantitativas_existentes:
    num_vars = len(var_quantitativas_existentes)
    num_cols = 3  # 3 colunas para melhor visualiza√ß√£o
    num_rows = ceil(num_vars / num_cols)
    
    # HISTOGRAMAS
    fig1, axes1 = plt.subplots(nrows=num_rows, ncols=num_cols, 
                              figsize=(15, 4 * num_rows))
    fig1.suptitle('üìà Distribui√ß√£o das Vari√°veis Quantitativas (Histogramas)', 
                  fontsize=16, fontweight='bold', y=1.05)
    
    # Garantir que axes1 seja sempre uma lista
    if num_vars == 1:
        axes1 = [axes1]
    elif num_rows == 1:
        axes1 = axes1.reshape(1, -1)
    
    axes1 = axes1.flatten()
    
    for i, col in enumerate(var_quantitativas_existentes):
        try:
            data_clean = df[col].dropna()
            
            if len(data_clean) > 0:
                sns.histplot(data=data_clean, ax=axes1[i], kde=True, 
                           alpha=0.7, color='skyblue')
                axes1[i].set_title(f'{col}', fontweight='bold', fontsize=11)
                axes1[i].grid(True, alpha=0.3)
                
                # Adicionar linha da m√©dia
                mean_val = data_clean.mean()
                axes1[i].axvline(mean_val, color='red', linestyle='--', 
                               linewidth=2, label=f'M√©dia: {mean_val:.1f}')
                axes1[i].legend(fontsize=9)
            else:
                axes1[i].text(0.5, 0.5, f'Sem dados\n{col}', 
                            ha='center', va='center', transform=axes1[i].transAxes)
        except Exception as e:
            axes1[i].text(0.5, 0.5, f'Erro: {col}', 
                        ha='center', va='center', transform=axes1[i].transAxes)
    
    # Remover subplots vazios
    for j in range(len(var_quantitativas_existentes), len(axes1)):
        axes1[j].remove()
    
    plt.tight_layout()
    plt.show()
    
    # BOXPLOTS
    fig2, axes2 = plt.subplots(nrows=num_rows, ncols=num_cols, 
                              figsize=(15, 4 * num_rows))
    fig2.suptitle('üì¶ Detec√ß√£o de Outliers (Boxplots)', 
                  fontsize=16, fontweight='bold', y=1.05)
    
    # Garantir que axes2 seja sempre uma lista
    if num_vars == 1:
        axes2 = [axes2]
    elif num_rows == 1:
        axes2 = axes2.reshape(1, -1)
    
    axes2 = axes2.flatten()
    
    for i, col in enumerate(var_quantitativas_existentes):
        try:
            data_clean = df[col].dropna()
            
            if len(data_clean) > 0:
                sns.boxplot(y=data_clean, ax=axes2[i], color='lightgreen')
                axes2[i].set_title(f'{col}', fontweight='bold', fontsize=11)
                axes2[i].grid(True, alpha=0.3)
                
                # Contar outliers
                Q1 = data_clean.quantile(0.25)
                Q3 = data_clean.quantile(0.75)
                IQR = Q3 - Q1
                outliers = data_clean[(data_clean < Q1 - 1.5*IQR) | 
                                    (data_clean > Q3 + 1.5*IQR)]
                
                if len(outliers) > 0:
                    axes2[i].text(0.02, 0.98, f'Outliers: {len(outliers)}', 
                                transform=axes2[i].transAxes, 
                                verticalalignment='top', fontsize=9,
                                bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))
            else:
                axes2[i].text(0.5, 0.5, f'Sem dados\n{col}', 
                            ha='center', va='center', transform=axes2[i].transAxes)
        except Exception as e:
            axes2[i].text(0.5, 0.5, f'Erro: {col}', 
                        ha='center', va='center', transform=axes2[i].transAxes)
    
    # Remover subplots vazios
    for j in range(len(var_quantitativas_existentes), len(axes2)):
        axes2[j].remove()
    
    plt.tight_layout()
    plt.show()

# =============================================================================
# 4. VISUALIZA√á√ïES DAS VARI√ÅVEIS QUALITATIVAS
# =============================================================================
print("\nüìä GERANDO VISUALIZA√á√ïES PARA VARI√ÅVEIS QUALITATIVAS...")

# Verifica se as colunas existem no DataFrame
var_qualitativas_existentes = [col for col in var_qualitativas if col in df.columns]

if var_qualitativas_existentes:
    num_vars_quali = len(var_qualitativas_existentes)
    num_cols_quali = 2  # 2 colunas para vari√°veis qualitativas
    num_rows_quali = ceil(num_vars_quali / num_cols_quali)
    
    fig3, axes3 = plt.subplots(nrows=num_rows_quali, ncols=num_cols_quali, 
                              figsize=(14, 5 * num_rows_quali))
    fig3.suptitle('üìä Distribui√ß√£o das Vari√°veis Qualitativas', 

                  fontsize=16, fontweight='bold', y=1.08)
    
    # Garantir que axes3 seja sempre uma lista
    if num_vars_quali == 1:
        axes3 = [axes3]
    elif num_rows_quali == 1:
        axes3 = axes3.reshape(1, -1)
    
    axes3 = axes3.flatten()
    
    for i, col in enumerate(var_qualitativas_existentes):
        try:
            # Contar valores √∫nicos
            value_counts = df[col].value_counts()
            
            if len(value_counts) > 0:
                # Limitar a 15 categorias mais frequentes
                if len(value_counts) > 15:
                    value_counts = value_counts.head(15)
                    title_suffix = f' (Top 15 de {len(df[col].unique())})'
                else:
                    title_suffix = ''
                
                # Criar gr√°fico de barras
                bars = axes3[i].bar(range(len(value_counts)), value_counts.values, 
                                  color='steelblue', alpha=0.8)
                
                axes3[i].set_title(f'{col}{title_suffix}', fontweight='bold', fontsize=12)
                axes3[i].set_xlabel(col, fontsize=10)
                axes3[i].set_ylabel('Frequ√™ncia', fontsize=10)
                axes3[i].set_xticks(range(len(value_counts)))
                axes3[i].set_xticklabels(value_counts.index, rotation=45, ha='right')
                axes3[i].grid(True, alpha=0.3)
                
                # Adicionar valores nas barras
                for bar, value in zip(bars, value_counts.values):
                    height = bar.get_height()
                    axes3[i].text(bar.get_x() + bar.get_width()/2., height,
                                f'{value}', ha='center', va='bottom', fontsize=9)
            else:
                axes3[i].text(0.5, 0.5, f'Sem dados\n{col}', 
                            ha='center', va='center', transform=axes3[i].transAxes)
        except Exception as e:
            axes3[i].text(0.5, 0.5, f'Erro: {col}', 
                        ha='center', va='center', transform=axes3[i].transAxes)
    
    # Remover subplots vazios
    for j in range(len(var_qualitativas_existentes), len(axes3)):
        axes3[j].remove()
    
    plt.tight_layout()
    plt.show()

else:
    print("‚ùå Nenhuma vari√°vel qualitativa encontrada no DataFrame!")

print("\n‚úÖ AN√ÅLISE EXPLORAT√ìRIA CONCLU√çDA!")
print("=" * 60)
```

```python
import pandas as pd
import numpy as np

# Definindo vari√°veis quantitativas e qualitativas com base nos dados fornecidos
var_quantitativas = ['Jan_Demand', 'Feb_Demand', 'Mar_Demand', 'Apr_Demand', 'May_Demand',
                     'Jun_Demand', 'Jul_Demand', 'Aug_Demand', 'Sep_Demand', 'Oct_Demand',
                     'Nov_Demand', 'Dec_Demand', 'Total_Annual_Units', 'Price_Per_Unit',
                     'Total_Sales_Value']
var_qualitativas = ['Item_ID', 'Item_Name', 'Category']

# Fun√ß√£o para detectar outliers usando IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][[column, 'Item_Name', 'Category']]
    return outliers

# 1. Identifica√ß√£o de Outliers nas Vari√°veis Quantitativas
print("üìä EXEMPLO DE OUTLIERS NAS VARI√ÅVEIS QUANTITATIVAS")
print("=" * 60)

for var in var_quantitativas:
    if var in df.columns:
        outliers = detect_outliers(df, var)
        if not outliers.empty:
            print(f"\nüîç {var}:")
            print(f"   ‚Ä¢ N√∫mero de outliers: {len(outliers)}")
            print(f"   ‚Ä¢ Exemplos de outliers:")
            print(outliers.head(3))  # Mostra os primeiros 3 exemplos
        else:
            print(f"\n‚úÖ {var}: Nenhum outlier detectado.")

# 2. Verifica√ß√£o de Inconsist√™ncias nas Vari√°veis Qualitativas
print("\n\nüîç VERIFICA√á√ÉO DE INCONSIST√äNCIAS NAS VARI√ÅVEIS QUALITATIVAS")
print("=" * 60)

for var in var_qualitativas:
    if var in df.columns:
        # Criar uma c√≥pia da coluna para an√°lise
        temp_col = df[var].astype(str).str.lower().str.strip()  # Normaliza para min√∫sculas e remove espa√ßos
        
        # Contar valores √∫nicos ap√≥s normaliza√ß√£o
        unique_normalized = temp_col.unique()
        unique_original = df[var].unique()
        
        # Verificar se h√° duplicatas disfar√ßadas (mesmo valor normalizado, mas diferente no original)
        potential_inconsistencies = {}
        for val_norm in unique_normalized:
            original_vals = df[var][temp_col == val_norm].unique()
            if len(original_vals) > 1:
                potential_inconsistencies[val_norm] = original_vals.tolist()
        
        print(f"\nüìã {var}:")
        print(f"   ‚Ä¢ Valores √∫nicos: {len(unique_original)}")
        print(f"   ‚Ä¢ Valores nulos: {df[var].isnull().sum()}")
        
        if potential_inconsistencies:
            print(f"   ‚Ä¢ Poss√≠veis inconsist√™ncias detectadas:")
            for norm_val, orig_vals in potential_inconsistencies.items():
                print(f"     - Normalizado: '{norm_val}' -> Originais: {orig_vals}")
        else:
            print(f"   ‚úÖ Nenhuma inconsist√™ncia √≥bvia detectada (ex.: mai√∫sculas/min√∫sculas ou espa√ßos).")

# Exibir amostra de dados com poss√≠veis inconsist√™ncias (se existirem)
if potential_inconsistencies:
    print("\nüìã AMOSTRA DE DADOS COM INCONSIST√äNCIAS:")
    for var in var_qualitativas:
        temp_col = df[var].astype(str).str.lower().str.strip()
        for norm_val, orig_vals in potential_inconsistencies.items():
            if norm_val in temp_col.values:
                sample = df[df[var].isin(orig_vals)].head(3)
                print(f"\nüîç {var} (Normalizado: '{norm_val}'):")
                print(sample[['Item_ID', var, 'Category']])
```

```python
import pandas as pd
import numpy as np

# Definindo vari√°veis quantitativas e qualitativas
var_quantitativas = ['Jan_Demand', 'Feb_Demand', 'Mar_Demand', 'Apr_Demand', 'May_Demand',
                     'Jun_Demand', 'Jul_Demand', 'Aug_Demand', 'Sep_Demand', 'Oct_Demand',
                     'Nov_Demand', 'Dec_Demand', 'Total_Annual_Units', 'Price_Per_Unit',
                     'Total_Sales_Value']
var_qualitativas = ['Item_ID', 'Item_Name', 'Category']

# 1. Tratamento de Dados Faltantes
print("üìä PREPARA√á√ÉO DOS DADOS")
print("=" * 60)
print("üîç 1. Verifica√ß√£o e Tratamento de Dados Faltantes:")
missing_data = df.isnull().sum()
if missing_data.sum() > 0:
    print("‚ö†Ô∏è Dados faltantes detectados e tratados:")
    for col in missing_data[missing_data > 0].index:
        if col in var_quantitativas:
            df[col].fillna(df[col].mean(), inplace=True)  # Substitui por m√©dia para num√©ricas
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)  # Substitui por moda para categ√≥ricas
    print(missing_data[missing_data > 0])
else:
    print("‚úÖ Nenhum dado faltante detectado. Nenhum tratamento realizado.")

# 2. Tratamento de Outliers
print("\nüîç 2. Tratamento de Outliers:")
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_count = len(df[(df[column] < lower_bound) | (df[column] > upper_bound)])
    df[column] = df[column].clip(lower_bound, upper_bound)  # Capa os valores
    return outliers_count

outlier_counts = {}
for var in var_quantitativas:
    if var in df.columns:
        count = cap_outliers(df, var)
        if count > 0:
            outlier_counts[var] = count

if outlier_counts:
    print("‚ö†Ô∏è Outliers tratados por capping (valores limitados ao IQR):")
    for var, count in outlier_counts.items():
        print(f"   ‚Ä¢ {var}: {count} outliers")
else:
    print("‚úÖ Nenhum outlier significativo detectado ou tratado.")

# 3. Remo√ß√£o de Linhas Duplicadas
print("\nüîç 3. Remo√ß√£o de Linhas Duplicadas:")
initial_rows = len(df)
df.drop_duplicates(inplace=True)
final_rows = len(df)
if initial_rows > final_rows:
    print(f"‚ö†Ô∏è {initial_rows - final_rows} linhas duplicadas removidas.")
else:
    print("‚úÖ Nenhuma linha duplicada detectada. Nenhum tratamento realizado.")

# 4. Padroniza√ß√£o de Categorias
print("\nüîç 4. Padroniza√ß√£o de Categorias:")
inconsistencies = {}
for var in var_qualitativas:
    if var in df.columns:
        temp_col = df[var].astype(str).str.lower().str.strip()
        unique_normalized = temp_col.unique()
        unique_original = df[var].unique()
        for val_norm in unique_normalized:
            original_vals = df[var][temp_col == val_norm].unique()
            if len(original_vals) > 1:
                inconsistencies[var] = {val_norm: original_vals.tolist()}

if inconsistencies:
    print("‚ö†Ô∏è Inconsist√™ncias detectadas e padronizadas:")
    for var, issues in inconsistencies.items():
        for norm_val, orig_vals in issues.items():
            df[var] = df[var].replace(orig_vals, orig_vals[0])  # Substitui por primeiro valor
            print(f"   ‚Ä¢ {var}: Normalizado '{norm_val}' -> '{orig_vals[0]}' (substituiu {orig_vals})")
else:
    print("‚úÖ Nenhuma inconsist√™ncia de categoria detectada. Nenhum tratamento realizado.")

# 5. Cria√ß√£o de Vari√°veis Derivadas
print("\nüîç 5. Cria√ß√£o de Vari√°veis Derivadas:")
df['Average_Monthly_Demand'] = df[['Jan_Demand', 'Feb_Demand', 'Mar_Demand', 'Apr_Demand',
                                   'May_Demand', 'Jun_Demand', 'Jul_Demand', 'Aug_Demand',
                                   'Sep_Demand', 'Oct_Demand', 'Nov_Demand', 'Dec_Demand']].mean(axis=1)
df['Sales_Per_Unit'] = df['Total_Sales_Value'] / df['Total_Annual_Units']
print("‚úÖ Criadas vari√°veis derivadas:")
print("   ‚Ä¢ Average_Monthly_Demand: M√©dia mensal de demanda.")
print("   ‚Ä¢ Sales_Per_Unit: Rela√ß√£o entre Total_Sales_Value e Total_Annual_Units.")

# 6. Valida√ß√£o das Transforma√ß√µes
print("\nüîç 6. Valida√ß√£o das Transforma√ß√µes:")
print("\nüìã Estat√≠sticas Descritivas Ap√≥s Transforma√ß√µes:")
print(df[var_quantitativas + ['Average_Monthly_Demand', 'Sales_Per_Unit']].describe().round(2))

print("\nüîç Verifica√ß√£o de Outliers Ap√≥s Tratamento:")
for var in var_quantitativas + ['Average_Monthly_Demand', 'Sales_Per_Unit']:
    if var in df.columns:
        outliers_post = detect_outliers(df, var)
        if not outliers_post.empty:
            print(f"‚ö†Ô∏è {var}: {len(outliers_post)} outliers restantes ap√≥s tratamento.")
        else:
            print(f"‚úÖ {var}: Nenhum outlier detectado ap√≥s tratamento.")

print("\n‚úÖ PREPARA√á√ÉO DOS DADOS CONCLU√çDA!")
print("=" * 60)

# Fun√ß√£o auxiliar para detectar outliers (usada na valida√ß√£o)
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]
```

```python
# Fun√ß√£o para criar tabelas de frequ√™ncia para vari√°veis qualitativas
def frequency_tables_qualitative(df, var_qualitativas):
    for var in var_qualitativas:
        if var in df.columns:
            # Frequ√™ncia Absoluta
            freq_abs = df[var].value_counts().sort_index()
            
            # Frequ√™ncia Relativa
            freq_rel = df[var].value_counts(normalize=True).sort_index().round(4) * 100
            
            # Frequ√™ncia Acumulada em Percentual
            freq_acum = (df[var].value_counts(normalize=True).sort_index().cumsum() * 100).round(2)
            
            print(f"\nüìä Tabelas de Frequ√™ncia para {var}")
            print("=" * 50)
            print(f"üî¢ Frequ√™ncia Absoluta:\n{freq_abs}\n")
            print(f"üìä Frequ√™ncia Relativa (%):\n{freq_rel}\n")
            print(f"üìà Frequ√™ncia Acumulada (%):\n{freq_acum}\n")

# Fun√ß√£o para criar faixas de valores e tabelas de frequ√™ncia para vari√°veis quantitativas
def frequency_tables_quantitative(df, var_quantitativas):
    for var in var_quantitativas:
        if var in df.columns:
            # Calcular n√∫mero de faixas (~10% dos dados por faixa)
            n_bins = int(len(df) * 0.1)  # Aproximadamente 10% por faixa
            if n_bins < 5:  # M√≠nimo de 5 faixas para evitar subdivis√µes muito grosseiras
                n_bins = 5
            
            # Criar faixas com base no m√≠nimo e m√°ximo, ajustando amplitude
            min_val = df[var].min()
            max_val = df[var].max()
            bin_width = (max_val - min_val) / n_bins
            bins = [min_val + i * bin_width for i in range(n_bins + 1)]
            
            # Garantir que a √∫ltima faixa inclua o valor m√°ximo
            bins[-1] = max_val + 1
            
            # Criar r√≥tulos para as faixas
            labels = [f'{bins[i]:.1f} - {bins[i+1]:.1f}' for i in range(len(bins)-1)]
            
            # Categorizar os dados
            df[f'{var}_binned'] = pd.cut(df[var], bins=bins, labels=labels, include_lowest=True)
            
            # Frequ√™ncia Absoluta
            freq_abs = df[f'{var}_binned'].value_counts().sort_index()
            
            # Frequ√™ncia Relativa
            freq_rel = df[f'{var}_binned'].value_counts(normalize=True).sort_index().round(4) * 100
            
            # Frequ√™ncia Acumulada em Percentual
            freq_acum = (df[f'{var}_binned'].value_counts(normalize=True).sort_index().cumsum() * 100).round(2)
            
            print(f"\nüìä Tabelas de Frequ√™ncia para {var}")
            print("=" * 50)
            print(f"üî¢ Frequ√™ncia Absoluta:\n{freq_abs}\n")
            print(f"üìä Frequ√™ncia Relativa (%):\n{freq_rel}\n")
            print(f"üìà Frequ√™ncia Acumulada (%):\n{freq_acum}\n")
            
            # Remover a coluna tempor√°ria ap√≥s uso
            df.drop(columns=[f'{var}_binned'], inplace=True)

# Executar as fun√ß√µes com as vari√°veis definidas anteriormente
print("üìä GERA√á√ÉO DE TABELAS DE FREQU√äNCIA")
print("=" * 60)

# Adicionar as novas vari√°veis derivadas
var_quantitativas.extend(['Average_Monthly_Demand', 'Sales_Per_Unit'])

# Gerar tabelas para vari√°veis qualitativas
frequency_tables_qualitative(df, var_qualitativas)

# Gerar tabelas para vari√°veis quantitativas
frequency_tables_quantitative(df, var_quantitativas)
```

```python
import pandas as pd
import numpy as np

def analise_abc_xyz(df):
    """
    Realiza an√°lise ABC-XYZ completa para gest√£o de estoques
    
    Args:
        df: DataFrame com dados de demanda e vendas
    
    Returns:
        DataFrame com classifica√ß√µes ABC-XYZ
    """
    
    if df is None or df.empty:
        raise ValueError("DataFrame est√° vazio ou n√£o foi carregado")
    
    print("Iniciando an√°lise ABC-XYZ...")
    print(f"Total de itens: {len(df)}")
    
    demand_cols = ['Jan_Demand', 'Feb_Demand', 'Mar_Demand', 'Apr_Demand', 
                   'May_Demand', 'Jun_Demand', 'Jul_Demand', 'Aug_Demand',
                   'Sep_Demand', 'Oct_Demand', 'Nov_Demand', 'Dec_Demand']
    
    missing_cols = [col for col in demand_cols if col not in df.columns]
    if missing_cols:
        print(f"Aviso: Colunas n√£o encontradas: {missing_cols}")
        demand_cols = [col for col in demand_cols if col in df.columns]
    
    if 'Total_Sales_Value' not in df.columns:
        print("Aviso: Coluna 'Total_Sales_Value' n√£o encontrada. Tentando alternativas...")
        value_cols = [col for col in df.columns if 'value' in col.lower() or 'sales' in col.lower()]
        if value_cols:
            print(f"Usando coluna: {value_cols[0]}")
            df['Total_Sales_Value'] = df[value_cols[0]]
        else:
            raise ValueError("Nenhuma coluna de valor de vendas encontrada")
    
    if demand_cols:
        df[demand_cols] = df[demand_cols].fillna(0)
        df['demanda_media_mensal'] = df[demand_cols].mean(axis=1)
        df['std_demanda'] = df[demand_cols].std(axis=1)
        df['cv_demanda'] = np.where(df['demanda_media_mensal'] > 0, 
                                   df['std_demanda'] / df['demanda_media_mensal'], 
                                   0)
    else:
        print("Aviso: Usando valores padr√£o para m√©tricas de demanda")
        df['demanda_media_mensal'] = 1
        df['std_demanda'] = 0.1
        df['cv_demanda'] = 0.1
    
    df_abc = df.copy().sort_values('Total_Sales_Value', ascending=False)
    df_abc['valor_cumulativo'] = df_abc['Total_Sales_Value'].cumsum()
    valor_total = df_abc['Total_Sales_Value'].sum()
    
    if valor_total > 0:
        df_abc['percentual_cumulativo'] = df_abc['valor_cumulativo'] / valor_total
        df_abc['classe_abc'] = np.where(df_abc['percentual_cumulativo'] <= 0.8, 'A',
                                       np.where(df_abc['percentual_cumulativo'] <= 0.95, 'B', 'C'))
    else:
        df_abc['classe_abc'] = 'C'
    
    df['classe_xyz'] = np.where(df['cv_demanda'] <= 0.25, 'X',
                               np.where(df['cv_demanda'] <= 0.5, 'Y', 'Z'))
    
    if 'Item_Name' not in df.columns:
        df['Item_Name'] = df['Item_ID']
    
    df_final = pd.merge(df_abc[['Item_ID', 'Item_Name', 'classe_abc', 'Total_Sales_Value', 'percentual_cumulativo']], 
                       df[['Item_ID', 'classe_xyz', 'cv_demanda', 'demanda_media_mensal']], 
                       on='Item_ID', how='inner')
    
    df_final['abc_xyz'] = df_final['classe_abc'] + df_final['classe_xyz']
    
    return df_final

def simular_economia_abc_xyz(df_final, custo_armazenagem_anual=0.15, custo_transporte_por_unidade=5.0, 
                            taxa_ruptura_cz=0.05, taxa_ruptura_ax=0.01, aumento_vendas_ax=0.1):
    """
    Simula cen√°rios para estimar o potencial de economia ao realocar recursos para itens AX.
    
    Args:
        df_final: DataFrame com resultados da an√°lise ABC-XYZ
        custo_armazenagem_anual: Custo anual de armazenagem como % do valor do estoque
        custo_transporte_por_unidade: Custo de transporte por unidade
        taxa_ruptura_cz: Taxa de ruptura estimada para itens CZ
        taxa_ruptura_ax: Taxa de ruptura estimada para itens AX
        aumento_vendas_ax: Aumento esperado nas vendas de itens AX com maior disponibilidade
    
    Returns:
        Dicion√°rio com resultados da simula√ß√£o
    """
    print("\nüìä SIMULA√á√ÉO DE CEN√ÅRIOS E CUSTO DE OPORTUNIDADE:")
    
    valor_total_estoque = df_final['Total_Sales_Value'].sum()
    valor_cz = df_final[df_final['abc_xyz'] == 'CZ']['Total_Sales_Value'].sum()
    valor_bz = df_final[df_final['abc_xyz'] == 'BZ']['Total_Sales_Value'].sum()
    valor_ax = df_final[df_final['abc_xyz'] == 'AX']['Total_Sales_Value'].sum()
    
    estoque_cz = valor_cz
    estoque_bz = valor_bz
    estoque_ax = valor_ax
    
    custo_armazenagem_base = valor_total_estoque * custo_armazenagem_anual
    custo_transporte_base = df_final['demanda_media_mensal'].sum() * 12 * custo_transporte_por_unidade
    custo_ruptura_cz = valor_cz * taxa_ruptura_cz
    custo_ruptura_ax = valor_ax * taxa_ruptura_ax
    custo_total_base = custo_armazenagem_base + custo_transporte_base + custo_ruptura_cz + custo_ruptura_ax
    
    reducao_estoque_cz = estoque_cz * 0.8
    reducao_estoque_bz = estoque_bz * 0.5
    capital_liberado = reducao_estoque_cz + reducao_estoque_bz
    novo_estoque_ax = estoque_ax + capital_liberado
    novo_valor_estoque = valor_total_estoque - reducao_estoque_cz - reducao_estoque_bz
    custo_armazenagem_otimizado = novo_valor_estoque * custo_armazenagem_anual
    
    demanda_cz = df_final[df_final['abc_xyz'] == 'CZ']['demanda_media_mensal'].sum() * 0.8
    demanda_bz = df_final[df_final['abc_xyz'] == 'BZ']['demanda_media_mensal'].sum() * 0.5
    nova_demanda_total = df_final['demanda_media_mensal'].sum() - demanda_cz - demanda_bz
    custo_transporte_otimizado = nova_demanda_total * 12 * custo_transporte_por_unidade
    
    novo_custo_ruptura_ax = novo_estoque_ax * taxa_ruptura_ax * 0.5
    novo_custo_ruptura_cz = (estoque_cz - reducao_estoque_cz) * taxa_ruptura_cz
    custo_total_otimizado = custo_armazenagem_otimizado + custo_transporte_otimizado + novo_custo_ruptura_ax + novo_custo_ruptura_cz
    
    aumento_vendas = novo_estoque_ax * aumento_vendas_ax
    economia_armazenagem = custo_armazenagem_base - custo_armazenagem_otimizado
    economia_transporte = custo_transporte_base - custo_transporte_otimizado
    economia_ruptura = (custo_ruptura_cz + custo_ruptura_ax) - (novo_custo_ruptura_cz + novo_custo_ruptura_ax)
    economia_total = economia_armazenagem + economia_transporte + economia_ruptura + aumento_vendas
    
    roi = (economia_total / capital_liberado) * 100 if capital_liberado > 0 else 0
    
    print("\nüìä CEN√ÅRIO BASE:")
    print(f"Valor total do estoque: R$ {valor_total_estoque:,.2f}")
    print(f"Custo de armazenagem: R$ {custo_armazenagem_base:,.2f}")
    print(f"Custo de transporte: R$ {custo_transporte_base:,.2f}")
    print(f"Custo de rupturas (CZ+AX): R$ {custo_ruptura_cz + custo_ruptura_ax:,.2f}")
    print(f"Custo total: R$ {custo_total_base:,.2f}")
    
    print("\nüìä CEN√ÅRIO OTIMIZADO (Reduzir CZ/BZ, investir em AX):")
    print(f"Capital liberado (CZ+BZ): R$ {capital_liberado:,.2f}")
    print(f"Novo valor do estoque: R$ {novo_valor_estoque:,.2f}")
    print(f"Novo custo de armazenagem: R$ {custo_armazenagem_otimizado:,.2f}")
    print(f"Novo custo de transporte: R$ {custo_transporte_otimizado:,.2f}")
    print(f"Novo custo de rupturas: R$ {novo_custo_ruptura_cz + novo_custo_ruptura_ax:,.2f}")
    print(f"Aumento de vendas (AX): R$ {aumento_vendas:,.2f}")
    print(f"Economia total: R$ {economia_total:,.2f}")
    print(f"ROI da realoca√ß√£o: {roi:.1f}%")
    
    return {
        'capital_liberado': capital_liberado,
        'economia_armazenagem': economia_armazenagem,
        'economia_transporte': economia_transporte,
        'economia_ruptura': economia_ruptura,
        'aumento_vendas': aumento_vendas,
        'economia_total': economia_total,
        'roi': roi
    }

def gerar_relatorio(df_final):
    """Gera relat√≥rio completo da an√°lise ABC-XYZ"""
    
    print("\n" + "="*60)
    print("RELAT√ìRIO DE AN√ÅLISE ABC-XYZ")
    print("="*60)
    
    # 1. Itens Classe A
    print("\n1. Quais itens representam maior valor financeiro e devem receber aten√ß√£o m√°xima (Classe A) :(Alto valor financeiro - Princ√≠pio 80/20):")
    print("-" * 50)
    itens_a = df_final[df_final['classe_abc'] == 'A'].sort_values('Total_Sales_Value', ascending=False)
    if not itens_a.empty:
        print(f"Total de itens A: {len(itens_a)}")
        print(f"Representam {len(itens_a)/len(df_final)*100:.1f}% dos itens")
        print(f"Valor total: R$ {itens_a['Total_Sales_Value'].sum():,.2f}")
        print("\nTop 10 itens classe A:")
        for idx, row in itens_a.head(10).iterrows():
            print(f"  {row['Item_ID']}: {row['Item_Name']} - R$ {row['Total_Sales_Value']:,.2f}")
    else:
        print("Nenhum item classe A encontrado")
    
    # 2. Itens Classe Z
    print("\n\n2. Quais itens t√™m consumo mais imprevis√≠vel e exigem buffers ou estrat√©gias diferenciadas (Classe Z):ITENS CLASSE Z (Demanda imprevis√≠vel - Alta variabilidade):")
    print("-" * 50)
    print("‚ÑπÔ∏è O Coeficiente de Varia√ß√£o (CV) mede a variabilidade da demanda em rela√ß√£o √† m√©dia. ")
    print("   - CV ‚â§ 0.25: Demanda est√°vel (Classe X)")
    print("   - 0.25 < CV ‚â§ 0.5: Demanda moderadamente vari√°vel (Classe Y)")
    print("   - CV > 0.5: Demanda altamente imprevis√≠vel (Classe Z)")
    print("   Um CV alto indica maior necessidade de estoques de seguran√ßa ou estrat√©gias como JIT.")
    
    itens_z = df_final[df_final['classe_xyz'] == 'Z'].sort_values('cv_demanda', ascending=False)
    if not itens_z.empty:
        print(f"\nTotal de itens Z: {len(itens_z)}")
        print(f"Representam {len(itens_z)/len(df_final)*100:.1f}% dos itens")
        print("\nTop 10 itens com maior variabilidade:")
        for idx, row in itens_z.head(10).iterrows():
            print(f"  {row['Item_ID']}: {row['Item_Name']} - CV: {row['cv_demanda']:.2f}")
    else:
        print("Nenhum item classe Z encontrado")
    
    # 3. Matriz ABC-XYZ
    print("\n\n3. Quais combina√ß√µes ABC-XYZ orientam pol√≠ticas espec√≠ficas de reabastecimento:MATRIZ ABC-XYZ E ESTRAT√âGIAS:")
    print("-" * 50)
    
    estrategias = {
        'AX': 'Revis√£o cont√≠nua - Estoque m√≠nimo (alta rotatividade + previs√≠vel)',
        'AY': 'Revis√£o peri√≥dica - Estoque moderado (alta rotatividade + m√©dia variabilidade)',
        'AZ': 'Revis√£o frequente - Alto estoque de seguran√ßa (alta rotatividade + imprevis√≠vel)',
        'BX': 'Pedido programado - Estoque controlado (m√©dia rotatividade + previs√≠vel)',
        'BY': 'Lote econ√¥mico - Revis√£o semanal (m√©dia rotatividade + m√©dia variabilidade)',
        'BZ': 'Estoque consignado ou JIT (m√©dia rotatividade + imprevis√≠vel)',
        'CX': 'Compra sob demanda (baixa rotatividade + previs√≠vel)',
        'CY': 'Pedidos agrupados (baixa rotatividade + m√©dia variabilidade)',
        'CZ': 'Considerar terceiriza√ß√£o ou elimina√ß√£o (baixa rotatividade + imprevis√≠vel)'
    }
    
    matriz_contagem = df_final['abc_xyz'].value_counts()
    
    for categoria in sorted(estrategias.keys()):
        qtd_itens = matriz_contagem.get(categoria, 0)
        print(f"\n{categoria}: {qtd_itens} itens ({qtd_itens/len(df_final)*100:.1f}%)")
        print(f"Estrat√©gia: {estrategias[categoria]}")
        if qtd_itens > 0:
            exemplos = df_final[df_final['abc_xyz'] == categoria].head(3)
            print("Exemplos:")
            for idx, row in exemplos.iterrows():
                print(f"  - {row['Item_ID']}: {row['Item_Name']}")
    
    # 4. Potencial de Economia
    print("\n\n4. Qual o potencial de economia ao redirecionar recursos para os grupos mais estrat√©gicos (ex: AX):POTENCIAL DE ECONOMIA AO REALOCAR RECURSOS PARA ITENS ESTRAT√âGICOS (AX):")
    print("-" * 50)
    resultados_simulacao = simular_economia_abc_xyz(df_final)
    
    print("\nüìå RESUMO DOS IMPACTOS:")
    print(f"Capital imobilizado liberado (CZ+BZ): R$ {resultados_simulacao['capital_liberado']:,.2f}")
    print(f"Redu√ß√£o de custos de armazenagem: R$ {resultados_simulacao['economia_armazenagem']:,.2f}")
    print(f"Redu√ß√£o de custos de transporte: R$ {resultados_simulacao['economia_transporte']:,.2f}")
    print(f"Redu√ß√£o de custos com rupturas: R$ {resultados_simulacao['economia_ruptura']:,.2f}")
    print(f"Aumento de vendas (AX): R$ {resultados_simulacao['aumento_vendas']:,.2f}")
    print(f"Economia total estimada: R$ {resultados_simulacao['economia_total']:,.2f}")
    print(f"ROI da realoca√ß√£o: {resultados_simulacao['roi']:.1f}%")
    
    # 5. Resumo Estat√≠stico e Recomenda√ß√µes
    print("\n\n5. Quais recomenda√ß√µes pr√°ticas podem ser entregues para as √°reas de compras, log√≠stica e financeiro?:")
  
    
    print("\nüìä RECOMENDA√á√ïES PR√ÅTICAS (5W2H):")
    print("As recomenda√ß√µes a seguir s√£o baseadas na an√°lise ABC-XYZ e na simula√ß√£o de cen√°rios, ")
    print("utilizando a t√©cnica 5W2H (What, Why, Where, When, Who, How, How Much) para orientar a√ß√µes.")
    print("Dashboards interativos podem ser criados para monitorar:")
    print("  - Distribui√ß√£o ABC-XYZ por categoria")
    print("  - Estoque ideal por grupo")
    print("  - KPIs de cobertura e rotatividade")
    
    ax_count = len(df_final[df_final['abc_xyz'] == 'AX'])
    ay_count = len(df_final[df_final['abc_xyz'] == 'AY'])
    az_count = len(df_final[df_final['abc_xyz'] == 'AZ'])
    cz_count = len(df_final[df_final['abc_xyz'] == 'CZ'])
    
    print("\nüì¶ COMPRAS:")
    print(f"  ‚Ä¢ O QUE: Focar negocia√ß√£o nos itens AX e AY ({ax_count + ay_count} itens).")
    print("  ‚Ä¢ POR QUE: Representam alto valor financeiro e maior impacto no faturamento.")
    print("  ‚Ä¢ ONDE: Nos contratos com fornecedores principais.")
    print("  ‚Ä¢ QUANDO: Imediatamente, revisando contratos trimestralmente.")
    print("  ‚Ä¢ QUEM: Equipe de compras e gerentes de suprimentos.")
    print(f"  ‚Ä¢ COMO: Estabelecer contratos de fornecimento para {ax_count} itens AX com prazos curtos.")
    print("  ‚Ä¢ QUANTO: Redu√ß√£o potencial de custos com negocia√ß√µes otimizadas.")
    print("  ‚Ä¢ A√á√ÉO: Negociar prazos mais flex√≠veis para itens BX e CX e revisar fornecedores de itens AZ.")
    
    print("\nüöö LOG√çSTICA:")
    print(f"  ‚Ä¢ O QUE: Priorizar localiza√ß√£o premium para {ax_count} itens AX.")
    print("  ‚Ä¢ POR QUE: Alta rotatividade exige acesso r√°pido para evitar rupturas.")
    print("  ‚Ä¢ ONDE: No armaz√©m principal, em √°reas de alta acessibilidade.")
    print("  ‚Ä¢ QUANDO: Durante o pr√≥ximo ciclo de reorganiza√ß√£o do armaz√©m.")
    print("  ‚Ä¢ QUEM: Equipe de log√≠stica e gestores de armaz√©m.")
    print(f"  ‚Ä¢ COMO: Criar buffers de seguran√ßa para {az_count} itens AZ e otimizar espa√ßo reduzindo √°rea para {cz_count} itens CZ.")
    print("  ‚Ä¢ QUANTO: Redu√ß√£o de custos log√≠sticos estimada em R$ {resultados_simulacao['economia_transporte']:,.2f}.")
    print("  ‚Ä¢ A√á√ÉO: Implementar picking diferenciado por categoria.")
    
    print("\nüí∞ FINANCEIRO:")
    print(f"  ‚Ä¢ O QUE: Redirecionar R$ {resultados_simulacao['capital_liberado']:,.2f} de CZ/BZ para AX.")
    print("  ‚Ä¢ POR QUE: Maximizar o ROI (estimado em {resultados_simulacao['roi']:.1f}%) e liberar capital imobilizado.")
    print("  ‚Ä¢ ONDE: No planejamento or√ßament√°rio e aloca√ß√£o de recursos.")
    print("  ‚Ä¢ QUANDO: No pr√≥ximo ciclo or√ßament√°rio.")
    print("  ‚Ä¢ QUEM: Equipe financeira e diretoria.")
    print("  ‚Ä¢ COMO: Monitorar KPIs de giro e cobertura via dashboards interativos.")
    print(f"  ‚Ä¢ QUANTO: Economia total estimada de R$ {resultados_simulacao['economia_total']:,.2f}.")
    print("  ‚Ä¢ A√á√ÉO: Estabelecer metas de redu√ß√£o de estoque para itens CZ e reinvestimento em AX.")

def main():
    """Fun√ß√£o principal para executar a an√°lise"""
    try:
        print("Carregando dados...")
        print("Gerando dados de exemplo para demonstra√ß√£o...")
        np.random.seed(42)
        n_items = 1000
        
        df = pd.DataFrame({
            'Item_ID': [f'ITEM_{i:04d}' for i in range(1, n_items + 1)],
            'Item_Name': [f'Produto {i}' for i in range(1, n_items + 1)],
            'Total_Sales_Value': np.random.lognormal(8, 1.5, n_items),
            'Jan_Demand': np.random.poisson(50, n_items),
            'Feb_Demand': np.random.poisson(45, n_items),
            'Mar_Demand': np.random.poisson(55, n_items),
            'Apr_Demand': np.random.poisson(60, n_items),
            'May_Demand': np.random.poisson(52, n_items),
            'Jun_Demand': np.random.poisson(48, n_items),
            'Jul_Demand': np.random.poisson(65, n_items),
            'Aug_Demand': np.random.poisson(58, n_items),
            'Sep_Demand': np.random.poisson(62, n_items),
            'Oct_Demand': np.random.poisson(70, n_items),
            'Nov_Demand': np.random.poisson(75, n_items),
            'Dec_Demand': np.random.poisson(80, n_items)
        })
        
        for i in range(0, n_items, 10):
            df.loc[i, ['Jul_Demand', 'Nov_Demand']] *= np.random.uniform(0.1, 3.0, 2)
        
        df_resultado = analise_abc_xyz(df)
        gerar_relatorio(df_resultado)
        nome_arquivo = 'analise_abc_xyz_resultado.xlsx'
        df_resultado.to_excel(nome_arquivo, index=False)
        print(f"\n‚úÖ An√°lise conclu√≠da! Resultado salvo em '{nome_arquivo}'")
        return df_resultado
        
    except Exception as e:
        print(f"‚ùå Erro durante a an√°lise: {str(e)}")
        return None

if __name__ == "__main__":
    resultado = main()
```
